```python
def get_content(url, base_url):
    """
    
    """

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Make the request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        all_categories = []
        all_menu_list = soup.find_all("div", class_="col-title col-title-special after")
        # Find all 'a' tags
        all_new_link_text = []
        if all_menu_list:
            for title_box in all_menu_list:
                link_box = title_box.find("a", class_="col-more pull-right")
                if link_box:
                    href = link_box.get('href')
                    href = urljoin(base_url, href)
                    
                    text = ""
                    title_name_box = title_box.find("h3", class_="title-word pull-left")
                    if title_name_box:
                        text = title_name_box.get_text(strip=True)
                    all_new_link_text.append([text, href])
            return all_new_link_text
        return None

    except Exception as e:
        print(e)
        return None
```

```python
def get_page_no(url):
    """
    
    """

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Make the request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        page_box = soup.find("div", class_="meneame")
        if page_box:
            all_page_link = page_box.find_all("a")
            if all_page_link:
                last_page = all_page_link[-1]
                last_page_link = last_page.get("href")
                
                href = urljoin(url, last_page_link)
                page_number = re.search(r'\d+', href)
                if page_number:
                    page_number.group()
                    return [href, page_number.group()]
                else:
                    return ["no link last page link" , 1]
        return ["no link last page link" , 1]
    except Exception as e:
        print(e)
        return ["no link last page link" , 1]

```

# updating all the page number issues

```python
def extract_index(url):
    # Pattern to match index_XX.html at the end of the URL
    pattern1 = r'index_(\d+)\.html$'
    
    # Pattern to match _X.html at the end of the URL
    pattern2 = r'_(\d+)\.html$'
    
    # Try to match the first pattern
    match = re.search(pattern1, url)
    if match:
        return match.group(1)
    
    # If first pattern doesn't match, try the second pattern
    match = re.search(pattern2, url)
    if match:
        return match.group(1)
    
    # If no pattern matches, return None
    return 1
```

```python
df.list_page_no = df.Last_page_link.apply(extract_index)
```
