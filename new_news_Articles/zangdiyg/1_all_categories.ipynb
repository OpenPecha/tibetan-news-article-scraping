```python
def get_content(url, base_url):
    """
    
    """

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Make the request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        # each_sub_cata = soup.find("div", style="background-color: rgba(243, 243, 243, 0); height: auto; line-height: 2; text-align: center;")
        
        # Find all 'a' tags within the div
        links = soup.find('div', style=lambda value: value and "background-color: rgba(243, 243, 243, 0)" in value).find_all('a')
        # Extract and print the links and text
        if links:
            all_new_link_text = []
            for link in links:
                href = link.get('href')
                text = link.h3.text.strip() if link.h3 else "No text found"
                print()
                href = base_url + href
                all_new_link_text.append([text, href])
            return all_new_link_text
        else:
            return None
          
    except Exception as e:
        print(e)

```

## Extarct each categories page number

```python
def get_page_no(url):
    """
    
    """

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Make the request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        page_last = soup.find("a", class_="end")
        if page_last:
            last_page_no = page_last.get_text(strip=True)
            return [page_last, last_page_no]
        else:
            all_page_link = soup.find_all("a", class_="num")
            if all_page_link:
                last_page_no = all_page_link[-1].get_text(strip=True)
                return [all_page_link[-1], last_page_no]
                    
        return ["no link last page link" , 0]
    except Exception as e:
        print(e)

```

```python
final_link = []
i = 0
for each_link in new_categories:
    link = each_link[1]
    page_no = get_page_no(link)
    clean_list = []
    clean_list.extend(each_link)
    clean_list.extend(page_no)
    # print(clean_list)
    final_link.append(clean_list)
    i = i+1
    if i%5 == 0:
        print(i)
```
