```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.tb1025.cn/lg/"
base_url = "https://www.tb1025.cn"
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
response = requests.get(url, headers=headers)
response.raise_for_status()

all_categories = []
# Parse the page content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
#  nav navbar-nav  nav-pills
catgory_center = soup.find("div", class_="collapse navbar-collapse")
if catgory_center:
    catgory_center = catgory_center.find("ul", class_="nav navbar-nav nav-pills")
    if catgory_center:
        # Find all 'a' tags
        links = catgory_center.find_all('a')
        # Extract and print the href attribute from each link
        for link in links:
            href = link.get('href')
            if href:
                href = base_url + href
                text = link.get_text(strip=True)
                all_categories.append((text, href))
```

```python
all_categories
```

```python
def get_content(url, base_url):
    """
    
    """

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Make the request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        # each_sub_cata = soup.find("div", style="background-color: rgba(243, 243, 243, 0); height: auto; line-height: 2; text-align: center;")
        
        # Find all 'a' tags within the div
        each_categories = soup.find_all('li', class_="title")

        # Extract and print the links and text
        if each_categories:
            all_new_link_text = []
            for each_cata in each_categories:
                text = each_cata.contents[0].strip()
                
                link_box = each_cata.find("a")
                href = link_box.get("href")
                href = base_url + href
                all_new_link_text.append([text, href])
            return all_new_link_text
        else:
            return None
          
    except Exception as e:
        print(e)
```

## extract sub categories

```python
i = 0
new_categories = []
base_url = "https://www.tb1025.cn"


for each_link in all_categories[1:]:
    print("\n")
    print(each_link)
    link = each_link[1]
    other_content = get_content(link, base_url)
    if other_content:
        print(other_content)
        new_categories.extend(other_content)
    else:
        new_categories.append(each_link)
```

## Extarct each categories page number

```python
def get_page_no(url):
    """
    
    """

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        # Make the request to the URL
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        # Parse the page content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        more_page = soup.find("li", class_="more page")
        if more_page:
            total_record = more_page.find("a", title="Total record")
            if total_record:
                all_page_link = more_page.find_all("a")
                if all_page_link:
                    link = all_page_link[-1].get("href")
                    
                    match = re.search(r'index_(\d+)\.html', link)
                    return [link, match.group(1)]
            return ["no link last page link" , 1]
        return ["no link last page link" , 1]
    except Exception as e:
        print(e)

```

```python
url = "https://www.tb1025.cn/dn/maoba/"
# url = "https://www.tb1025.cn/luoza/boju/index.html"

get_page_no(url)
```

```python
final_link = []
i = 0
for each_link in new_categories:
    link = each_link[1]
    page_no = get_page_no(link)
    clean_list = []
    clean_list.extend(each_link)
    clean_list.extend(page_no)
    # print(clean_list)
    final_link.append(clean_list)
    i = i+1
    if i%5 == 0:
        print(i)
```
