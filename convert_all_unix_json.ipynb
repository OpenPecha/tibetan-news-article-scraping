```python
import json
import os
```

```python
def read_json(file_path):
    """
    Read a JSON file and return its contents.
    """
    with open(file_path, 'r', encoding='utf-8') as openfile:
        loaded_file = json.load(openfile)
        print(f"Successfully loaded: {os.path.basename(file_path)}")
    return loaded_file

def save_json(file_path, data):
    """
    Save data to a JSON file with proper ASCII formatting.
    """
    with open(file_path, "w", encoding='utf-8') as outfile:
        json.dump(data, outfile, indent=4, ensure_ascii=False)
        print(f"Successfully saved: {os.path.basename(file_path)}")
```

```python

def process_json_files(directory):
    """
    Walk through the given directory, process all JSON files,
    and save them back with proper ASCII formatting.
    """
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                
                # Load the JSON file
                data = read_json(file_path)
                
                # Save the JSON file with proper ASCII formatting
                save_json(file_path, data)

```

```python
directory_path = "./"
process_json_files(directory_path)
```
